```python
import pandas as pd
df=pd.read_csv('medical_students_dataset.csv')

```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == "object":
            # Fill categorical with mode (most frequent value)
            df[col].fillna(df[col].mode()[0], inplace=True)
            print(f"Filled missing values in '{col}' with MODE: {df[col].mode()[0]}")
        else:
            # For numeric columns
            if df[col].skew() < 1:
                # If distribution is normal → use mean
                df[col].fillna(df[col].mean(), inplace=True)
                print(f"Filled missing values in '{col}' with MEAN: {df[col].mean()}")
            else:
                # If distribution is skewed → use median
                df[col].fillna(df[col].median(), inplace=True)
                print(f"Filled missing values in '{col}' with MEDIAN: {df[col].median()}")
```

```python
print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print("Number of duplicate rows after removal:", df.duplicated().sum())
```
